## Inference with a single image on custom pretrained detection model 

```python
import numpy as np
import matplotlib.pyplot as plt
from doctr.io import DocumentFile
from doctr.models import db_resnet50
import torch
import cv2

# Load model and set device
device = "cuda" if torch.cuda.is_available() else "cpu"
model_path = "D:/JIN/Capstone_Spring/doctr/checkpoints/db_resnet50_20250322-171811.pt"
det_model = db_resnet50(pretrained=False)
det_model.load_state_dict(torch.load(model_path, map_location="cpu"))
det_model.to(device)
det_model.eval()

# Process image
img_path = "D:/JIN/Capstone_Spring/doctr/rotated.png"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_resized = cv2.resize(img, (1024, 1024))

# Convert to tensor WITHOUT normalization
temp_path = "temp_resized.jpg" 
cv2.imwrite(temp_path, cv2.cvtColor(img_resized, cv2.COLOR_RGB2BGR))
doc = DocumentFile.from_images(temp_path)
doc_tensor = torch.from_numpy(doc[0]).type(torch.float32).permute(2, 0, 1) / 255.0 # No division by 255
doc_tensor = doc_tensor.unsqueeze(0).to(device)

# Get predictions
with torch.inference_mode():
    result = det_model(doc_tensor)

# Draw bounding boxes on the image
def draw_boxes(image, predictions, threshold=0.5):
    """Draw bounding boxes on the image based on model predictions."""
    img_with_boxes = image.copy()
    h, w = img_with_boxes.shape[:2]
    
    # Extract bounding boxes
    boxes = predictions['preds'][0]['words']
    
    # Draw each box that meets the confidence threshold
    for box in boxes:
        if box[4] >= threshold:  # 5th value is confidence score
            # Convert normalized coordinates to pixel coordinates
            xmin = int(box[0] * w)
            ymin = int(box[1] * h)
            xmax = int(box[2] * w)
            ymax = int(box[3] * h)
            
            # Draw the rectangle
            cv2.rectangle(img_with_boxes, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
            
            # Optionally, add the confidence score text
            confidence = f"{box[4]:.2f}"
            cv2.putText(img_with_boxes, confidence, (xmin, ymin-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
    return img_with_boxes

# Draw boxes with a confidence threshold of 0.6
img_with_boxes = draw_boxes(img_resized, result, threshold=0.1)

# Display the result
plt.figure(figsize=(12, 12))
plt.imshow(img_with_boxes)
plt.axis('off')
plt.show()

# Save the result
cv2.imwrite("result_with_boxes.jpg", cv2.cvtColor(img_with_boxes, cv2.COLOR_RGB2BGR))
# result
```

```python
from doctr.models.detection.predictor import DetectionPredictor
from doctr.models.recognition.predictor import RecognitionPredictor
from doctr.models.preprocessor import PreProcessor
from doctr.models.predictor import OCRPredictor
from doctr.models import vitstr_small, db_resnet50, ocr_predictor
from doctr.datasets import VOCABS
import torch
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
# from doctr.utils.visualization import visualize_page
import cv2
from rembg import remove

def preprocess_image(img_path, output_path=None, remove_bg=True, rotate=True):
    """
    Preprocess image for OCR
    1. Read image
    2. Remove background (optional)
    3. Rotate to correct orientation (optional)
    4. Resize if needed
    
    Args:
        img_path: Path to input image
        output_path: Path to save preprocessed image (optional)
        remove_bg: Whether to remove background
        rotate: Whether to rotate image to correct orientation
        
    Returns:
        Preprocessed image as numpy array
    """
    # Read image
    print(f"Reading image from {img_path}")
    image = cv2.imread(img_path)
    if image is None:
        raise ValueError(f"Could not read image from {img_path}")
    
    # Store original for visualization
    original_image = image.copy()
    
    # Remove background if requested
    if remove_bg:
        print("Removing background...")
        # Remove background
        mask = remove(image)
        # Print shape after background removal
        print(f"Shape after background removal: {mask.shape}")
        
        # Convert RGBA to RGB if the mask has 4 channels
        if mask.shape[2] == 4:
            print("Converting RGBA to RGB...")
            # Create a white background
            white_background = np.ones_like(mask, dtype=np.uint8) * 255
            # Only keep RGB channels from white_background (remove alpha)
            white_background = white_background[:,:,:3]
            
            # Extract alpha channel from mask
            alpha = mask[:,:,3:4] / 255.0  # Normalize alpha to 0-1
            
            # Extract RGB channels from mask
            rgb = mask[:,:,:3]
            
            # Blend RGB with white background using alpha
            blended = (rgb * alpha + white_background * (1 - alpha)).astype(np.uint8)
            
            image = blended
            print(f"Converted image shape: {image.shape}")
        else:
            image = mask

    # Rotate image if requested
    if rotate:
        print("Detecting and correcting orientation...")
        # Convert to grayscale for text line detection
        if len(image.shape) == 3:
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        else:
            gray = image
            
        # Apply thresholding
        ret, binary = cv2.threshold(gray, 80, 255, cv2.THRESH_BINARY_INV)
        
        # Apply dilation to connect characters into text lines
        kernel_width = 25  # Adjust as needed
        kernel = np.ones((3, kernel_width), np.uint8)
        dilated = cv2.dilate(binary, kernel, iterations=1)
        
        # Find contours to identify text lines
        contours, _ = cv2.findContours(dilated, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
        
        # Filter contours by area
        min_area = 1000
        min_area_rects = []
        
        for contour in contours:
            area = cv2.contourArea(contour)
            if area < min_area:
                continue
                
            # Find minimum area rectangle
            rect = cv2.minAreaRect(contour)
            min_area_rects.append(rect)
        
        # Calculate mean angle
        if min_area_rects:
            angles = []
            for rect in min_area_rects:
                center, (width, height), angle = rect
                # Adjust angle based on orientation
                if width < height:
                    adjusted_angle = angle - 90
                else:
                    adjusted_angle = angle
                angles.append(adjusted_angle)
            
            mean_angle = sum(angles) / len(angles)
            print(f"Calculated mean angle: {mean_angle:.2f} degrees")
            
            # Rotate image
            h, w = image.shape[:2]
            center = (w // 2, h // 2)
            rotation_matrix = cv2.getRotationMatrix2D(center, mean_angle, 1.0)
            
            # Calculate new dimensions
            cos = np.abs(rotation_matrix[0, 0])
            sin = np.abs(rotation_matrix[0, 1])
            new_w = int((h * sin) + (w * cos))
            new_h = int((h * cos) + (w * sin))
            
            # Adjust rotation matrix
            rotation_matrix[0, 2] += (new_w / 2) - center[0]
            rotation_matrix[1, 2] += (new_h / 2) - center[1]
            
            # Perform rotation
            image = cv2.warpAffine(image, rotation_matrix, (new_w, new_h), flags=cv2.INTER_LINEAR)
    
    # Resize if needed (optional, doctr preprocessor handles this)
    image = cv2.resize(image, (1024, 1024), interpolation=cv2.INTER_AREA)
    
    return image

def ocr_inference(image, predictor):
    """
    Run OCR inference on a preprocessed image
    
    Args:
        image: Preprocessed image (numpy array)
        predictor: OCR predictor
        
    Returns:
        OCR result
    """
    print("Running OCR inference...")
    
    # Check if image is numpy array
    if not isinstance(image, np.ndarray):
        raise ValueError("Image must be a numpy array")
    
    # Ensure image has 3 channels (RGB)
    if len(image.shape) == 2:
        # Convert grayscale to RGB
        print("Converting grayscale to RGB...")
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    elif len(image.shape) == 3 and image.shape[2] == 4:
        # Convert RGBA to RGB
        print("Converting RGBA to RGB...")
        image = cv2.cvtColor(image, cv2.COLOR_RGBA2RGB)
    elif len(image.shape) == 3 and image.shape[2] == 3:
        # Already RGB, make sure it's in the right order (OpenCV uses BGR)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    print(f"Image shape before inference: {image.shape}")
    
    # Run OCR prediction
    result = predictor([image])
    
    return result

# def visualize_result(image, result, output_path=None):
#     """
#     Visualize OCR result
    
#     Args:
#         image: Original image
#         result: OCR result
#         output_path: Path to save visualization
#     """
#     print("Visualizing OCR result...")
    
#     # Convert to RGB for visualization
#     if len(image.shape) == 3 and image.shape[2] == 3:
#         if image.dtype == np.uint8:
#             rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#         else:
#             rgb_image = image
#     else:
#         # Convert grayscale to RGB
#         rgb_image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    
#     # Visualize the result
#     plt.figure(figsize=(15, 10))
#     visualize_page(rgb_image, result.pages[0].blocks, result.pages[0].blocks, interactive=False)
    
#     # Save visualization if requested
#     if output_path:
#         plt.savefig(output_path)
#         print(f"Saved visualization to {output_path}")
    
#     # Show visualization
#     plt.show()
    
#     # Print detected text
#     print("\nDetected Text:")
#     for page in result.pages:
#         for block in page.blocks:
#             for line in block.lines:
#                 line_text = " ".join([word.value for word in line.words])
#                 print(line_text)


# Main execution
if __name__ == "__main__":
    # Load detection model
    det_checkpoint = "D:/JIN/Capstone_Spring/doctr/checkpoints/db_resnet50_20250320-224057.pt"
    det_model = db_resnet50(pretrained=False)
    det_model.load_state_dict(torch.load(det_checkpoint, map_location="cpu"))

    # Load recognition model
    reco_checkpoint = "D:/JIN/Capstone_Spring/doctr/checkpoints/vitstr_small_20250320-231021.pt"
    reco_model = vitstr_small(pretrained=False, vocab=VOCABS["vietnamese"])
    reco_model.load_state_dict(torch.load(reco_checkpoint, map_location="cpu"))

    # Combine both predictors
    predictor = ocr_predictor(det_model, reco_model, preserve_aspect_ratio=True)

    # Path to input image
    img_path = "C:/Users/Admin/Downloads/Label/Label/Images/IMG_2790.jpeg"
    
    # Path to save preprocessed image
    output_dir = "C:/Users/Admin/Downloads/Label/Label/Processed"
    import os
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # Get base filename
    base_filename = os.path.basename(img_path)
    preprocessed_path = os.path.join(output_dir, f"preprocessed_{base_filename}")
    visualization_path = os.path.join(output_dir, f"ocr_result_{os.path.splitext(base_filename)[0]}.png")

    # Preprocess image
    preprocessed_image = preprocess_image(
        img_path, 
        preprocessed_path,
        remove_bg=True,
        rotate=True
    )

    # Run OCR inference
    result = ocr_inference(preprocessed_image, predictor)

    # Visualize result
    # visualize_result(preprocessed_image, result, visualization_path)
    
    print("Processing complete!")
```
