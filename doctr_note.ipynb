## Inference with a single image on custom pretrained detection model 

```python
import numpy as np
import matplotlib.pyplot as plt
from doctr.io import DocumentFile
from doctr.models import db_resnet50
import torch
import cv2

# Load model and set device
device = "cuda" if torch.cuda.is_available() else "cpu"
model_path = ".\db_resnet50_20250318-032905.pt"
det_model = db_resnet50(pretrained=False)
det_model.load_state_dict(torch.load(model_path, map_location="cpu"))
det_model.to(device)
det_model.eval()

# Process image
img_path = "D:/JIN/Capstone_Spring/doctr/data/images/IMG_1860.jpeg"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_resized = cv2.resize(img, (1024, 1024))

# Convert to tensor WITHOUT normalization
temp_path = "temp_resized.jpg" 
cv2.imwrite(temp_path, cv2.cvtColor(img_resized, cv2.COLOR_RGB2BGR))
doc = DocumentFile.from_images(temp_path)
doc_tensor = torch.from_numpy(doc[0]).type(torch.float32).permute(2, 0, 1)  # No division by 255
doc_tensor = doc_tensor.unsqueeze(0).to(device)

# Get predictions
with torch.inference_mode():
    result = det_model(doc_tensor)

# Draw bounding boxes on the image
def draw_boxes(image, predictions, threshold=0.5):
    """Draw bounding boxes on the image based on model predictions."""
    img_with_boxes = image.copy()
    h, w = img_with_boxes.shape[:2]
    
    # Extract bounding boxes
    boxes = predictions['preds'][0]['words']
    
    # Draw each box that meets the confidence threshold
    for box in boxes:
        if box[4] >= threshold:  # 5th value is confidence score
            # Convert normalized coordinates to pixel coordinates
            xmin = int(box[0] * w)
            ymin = int(box[1] * h)
            xmax = int(box[2] * w)
            ymax = int(box[3] * h)
            
            # Draw the rectangle
            cv2.rectangle(img_with_boxes, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
            
            # Optionally, add the confidence score text
            confidence = f"{box[4]:.2f}"
            cv2.putText(img_with_boxes, confidence, (xmin, ymin-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
    return img_with_boxes

# Draw boxes with a confidence threshold of 0.6
img_with_boxes = draw_boxes(img_resized, result, threshold=0.75)

# Display the result
plt.figure(figsize=(12, 12))
plt.imshow(img_with_boxes)
plt.axis('off')
plt.show()

# Save the result
cv2.imwrite("result_with_boxes.jpg", cv2.cvtColor(img_with_boxes, cv2.COLOR_RGB2BGR))
```
